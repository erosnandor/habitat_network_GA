```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
reloc = pd.read_csv("p_reloc_proportions.csv")
insert = pd.read_csv("p_insert_proportions.csv")
elits = pd.read_csv("elit_proportions.csv")
elits.head()
```

```python
fig, ax = plt.subplots(3,figsize=(4.5,10))

for key, grp in reloc.groupby(['P_reloc']):
    ax[0].plot(grp['Generation_number'], grp['Fitness'], label=key)


for key, grp in insert.groupby(['P_insert']):
    ax[1].plot(grp['Generation_number'], grp['Fitness'], label=key)

for key, grp in elits.groupby(['Elit_prop']):
    ax[2].plot(grp['Generation_number'], grp['Fitness'], label=key)


ax[0].legend(fontsize=11, title = "PR:")
ax[0].set_title("a)", loc='left', fontsize=14)
ax[0].set_xlabel("Generation number", fontsize = 13)
ax[0].set_ylabel("Fitness value", fontsize = 13)

ax[1].legend(fontsize=11, title = "PI:")
ax[1].set_title("b)", loc='left', fontsize=14)
ax[1].set_xlabel("Generation number", fontsize = 13)
ax[1].set_ylabel("Fitness value", fontsize = 13)

ax[2].legend(fontsize=11, title = "PE:")
ax[2].set_title("c)", loc='left', fontsize=14)
ax[2].set_xlabel("Generation number", fontsize = 13)
ax[2].set_ylabel("Fitness value", fontsize = 13)

fig.subplots_adjust(hspace=0.3)
plt.savefig("Parameter_tuning.pdf", bbox_inches='tight')
plt.show()
```
